```python
import numpy as np
import pandas as pd

from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn import tree

import matplotlib.pyplot as plt
```

```python
# followers / friends
def ffr(x):
    if x[2] > 0:
        return x[1] / (x[2] * 1.0)
    return 0.0

# statuses / favorites
def stfv(x):
    if x[4] > 0:
        return x[3] / (x[4] * 1.0)
    return 0.0

# load training dataset
TRAINING_FILE = '../data/model.dat'
TRAINING_DATA = pd.read_csv(TRAINING_FILE)

# TRAINING_DATA['ffr'] = TRAINING_DATA.apply(ffr, axis=1)
# TRAINING_DATA['stfv'] = TRAINING_DATA.apply(stfv, axis=1)

TRAINING_DATA.head()
```

```python
# assign x, y
X = TRAINING_DATA.values[:, 1:]
Y = TRAINING_DATA.values[:, 0]

# assign classes, features
class_names = list(TRAINING_DATA['Class'].unique())
feature_names = list(TRAINING_DATA.columns.values)[1:]
```

```python
# init classifier
clf = RandomForestClassifier(n_jobs = -1, 
                            criterion = 'entropy', 
                            random_state = 42,
                            max_depth = 5, 
                            min_samples_leaf = 5, n_estimators = 8)
clf.fit(X, Y)
```

```python
UN = pd.read_csv('../../../www/twitterbots-server-node/data/sample.csv')
UN.head()
```

```python
UN_X = UN[['FollowersCount', 'FriendsCount', 'StatusesCount', 'FavouritesCount', 'ListedCount']]
UN_X.head()
```

```python
UN['Label'] = clf.predict(UN_X)
```
